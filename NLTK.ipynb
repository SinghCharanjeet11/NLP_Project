NLTK Natural Language Text Processing
4 things to do with NLTK
*   Tokenization
*   Stopwords List
*   Stemming
*   Lemmetization   

```python
import nltk
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
```

```python
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import  word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
raw_text="I am  learning Python programming, Python is greatest language..!!"
```

```python
# Step-1 : Lowercase
raw_text=raw_text.lower()
print("After lowercase: ",raw_text)
```

```python
# Step-2 Tokenization
print(raw_text.split())
```

```python
tokens=word_tokenize(raw_text)
print("Tokens:  ",tokens)
```

```python
# Step-3 Remove Stopwords
english_stopwords=stopwords.words("english")
print("Stopwords: ",english_stopwords)
```

```python
filtered_tokens=[]
for word in tokens:
  if word not in english_stopwords:
    filtered_tokens.append(word)
print("Filtered tokens: ",filtered_tokens)
```

```python
# Step-4 Remove Punctuations
import string
punctuations=string.punctuation
print("Punctuations Available: ",punctuations)
```

```python
clean_tokens=[word for word in filtered_tokens if word not in punctuations]
print("After removing punctuations: ", clean_tokens)
```

```python
stem=PorterStemmer()
stem.stem("playing")
stem.stem("greatest")
```

```python
wnet=WordNetLemmatizer()
wnet.lemmatize("greatest",pos="a")
```

```python
lemmatized_word=[]
for word in clean_tokens:
  lemmatized_word.append(wnet.lemmatize(word,"v"))
print("After lemmatizing: ",lemmatized_word)
```

```python
final_tokens=[]
for word in lemmatized_word:
  if word.isalpha():
    final_tokens.append(word)
print("Final Tokens: ",final_tokens)

```

```python
cleaned_text=" ".join(final_tokens)
print("Cleaned Text: ",final_tokens)
```
